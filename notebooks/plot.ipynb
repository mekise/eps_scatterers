```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as la
```

```python
data = np.load("../data/data.npz")
scattpos = np.load("../data/data.npz")["scattpos"]
```

```python
plt.figure(dpi=120)

for i in range(scattpos.shape[0]):
    plt.axvline(la.norm(scattpos[i, :]), color="gray", linewidth='1', linestyle='dashed')
plt.axhline(data["analyticalsum"], color="gray", linewidth='1', linestyle='dashed')
plt.plot(data["r"], data["P"], color="C1")

plt.xlabel("r")
plt.ylabel("P")
```
